# Libraries

```python
import pandas as pd
import numpy as np
import pickle
```

# Read and Merge files

```python
# Intialize year list and dataframe
year = list(range(2019, 2023))
df = pd.DataFrame()
cognate_search = pd.DataFrame()

# concatenate dataframes
for y in year:
    globals()["df_" + str(y)] = pd.read_csv("/Volumes/Elements/CF_dataset 2/" + "df_" + str(y) + ".csv")  # read files
    df = pd.concat([df, globals()["df_" + str(y)]])  # concat
    
    globals()["cognate_search_" + str(y)] = pd.read_csv("/Volumes/Elements/CF_dataset 2/" + "cognate_search_" + str(y) + ".csv")  # read files
    cognate_search = pd.concat([cognate_search, globals()["cognate_search_" + str(y)]])  # concat

    
```

```python
# View dataframe
display(df.head())
#display(df.info(verbose = True, null_counts= True))

display(cognate_search.head())
#display(cognate_search.info(verbose = True, null_counts = True))
```

# Export files: pickle and csv

```python
# To csv
df.to_csv("/Volumes/Elements/CF_dataset 2/data_master.csv")
cognate_search.to_csv("/Volumes/Elements/CF_dataset 2/cognate_search.csv")

# use pickle to store
object_filename = "/Volumes/Elements/CF_dataset 2/df.sav"

saved_model = pickle.dump(df, open(object_filename,'wb'))

print('Dataframe is saved into to disk successfully Using Pickle')

object_filename = "/Volumes/Elements/CF_dataset 2/cognate_search.sav"

saved_model = pickle.dump(df, open(object_filename,'wb'))

print('Dataframe is saved into to disk successfully Using Pickle')


# # Code to retrieve pickle
# # use pickle to retrieve

# file = open("df.sav",'rb')
# df_corr = pickle.load(file)
# file.close()
```
