# Preprocessing

```python
df["plabel"] = df["plabel"].apply(lambda x: x.lower())  #lower
df["plabel"] = df["plabel"].apply(lambda x: x.strip())  #remove spaces
df["plabel"] = df["plabel"].apply(
    lambda x: re.compile('<.*?>').sub(' ', x))  # remove punctuations
df["plabel"] = df["plabel"].apply(
    lambda x: re.compile('[%s]' % re.escape(string.punctuation)).sub(' ', x))
df["plabel"] = df["plabel"].apply(
    lambda x: re.sub('\s+', ' ', x))  # remove spaces
df["plabel"] = df["plabel"].apply(
    lambda x: re.sub(r'\[[0-9]*\]', ' ', x))  # remove digits
df["plabel"] = df["plabel"].apply(lambda x: re.sub(r'[^\w\s]', '',
                                                   str(x).lower().strip()))
df["plabel"] = df["plabel"].apply(lambda x: re.sub(r'\d', ' ', x))
df["plabel"] = df["plabel"].apply(lambda x: re.sub(r'\s+', ' ', x))
```

```python
# create functions
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use', "jbe"]) # extend stopwords

# remove punctuation
def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(simple_preprocess(str(sentence), deacc=True))

        
# # remove stopwords
# def remove_stopwords(texts):
#     return [[word for word in simple_preprocess(str(doc)) 
#              if word not in stop_words] for doc in texts]
# create bigrams
def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

# create trigrams
def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

#stemming
def stemming(text):
    texts_out = []
    for sent in text:
        texts_out.append([stemmer.stem(y) for y in sent])
    return texts_out

#Create sentence again for dataframe updation
def word_to_sent(text):
    texts_out = []
    for word in text:
        texts_out.append(" ".join(y for y in word))
    return texts_out
```

```python
%%time

# Preprocess plabel

# data words
data = df.plabel.values.tolist()
data_words_plabel = list(sent_to_words(data))

# bigram and trigrams phraser
# Build the bigram and trigram models
bigram = gensim.models.Phrases(data_words_plabel, min_count=2, threshold=10) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_words_plabel], threshold=10)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

# # remove stop words
# data_words_plabel = remove_stopwords(data_words_plabel)

# form bigrams
data_words_plabel = make_bigrams(data_words_plabel)

# # Do stemming
# # Use English stemmer.
# stemmer = SnowballStemmer("english")
# data_words_plabel = stemming(data_words_plabel)

# Get data in sentence form
data_sent_plabel = word_to_sent(data_words_plabel)
```

```python
%%time

## get class_weights
classes = np.unique(y_train)
weights = class_weight.compute_sample_weight(class_weight='balanced', y=y_train)
classes_weights = dict(zip(classes, weights))
    
## get model list
models = [MultinomialNB(), DecisionTreeClassifier(), RandomForestClassifier(), XGBClassifier(),
          AdaBoostClassifier(), SVC(), CatBoostClassifier(), LGBMClassifier(), MLPClassifier(), 
          MultinomialNB(), DecisionTreeClassifier(class_weight = "balanced"), RandomForestClassifier(class_weight = "balanced"),
          RandomForestClassifier(class_weight = "balanced_subsample"), XGBClassifier(), AdaBoostClassifier(), 
          SVC(class_weight = "balanced"), CatBoostClassifier(class_weights = classes_weights), 
          LGBMClassifier(class_weight = "balanced")]

## CV dataframe. splits = 5 set before stepwise modeling
splits = 3

CV = StratifiedKFold(n_splits=splits)


globals()["cv_df_tag_"+str(current_disagg)] = pd.DataFrame(index=range(splits * len(models)))

## update in loop
entries = []
for model in models:
    model_name = model.__class__.__name__

    if (("class_weight", None) not in model.get_params().items()) and (any(model_name in sl for sl in entries)):
        if "class_weight" in model.get_params().keys():
            ## modify name
            weight_method = model.get_params()["class_weight"]
            model_name = model_name + " " + weight_method
            print(model_name)
            
            ## cross val score
            accuracies = cross_val_score(model,
                                 X_train_modeling,
                                 y_train,
                                 scoring='accuracy',
                                 cv=CV)

        elif "class_weights" in model.get_params().keys():
            ## modify name
            weight_method = "balanced class weights"
            model_name = model_name + " " + weight_method
            print(model_name)
            
            ## cross val score
            accuracies = cross_val_score(model,
                                 X_train_modeling,
                                 y_train,
                                 scoring='accuracy',
                                 cv=CV)

        else:
            ## modify name
            weight_method = "balanced sample weights"
            model_name = model_name + " " + weight_method
            print(model_name)
            
            ## cross val score
            fit_parameters = {"sample_weight": weights}
            accuracies = cross_val_score(model,
                                 X_train_modeling,
                                 y_train,
                                 scoring='accuracy',
                                 cv=CV, fit_params=fit_parameters)
            
    else:
        print(model_name)
        accuracies = cross_val_score(model,
                                 X_train_modeling,
                                 y_train,
                                 scoring='accuracy',
                                 cv=CV)
    
    # append entries for dataframe
    for fold_idx, accuracy in enumerate(accuracies):
        entries.append((model_name, fold_idx, accuracy))
        
globals()["cv_df_tag_"+str(current_disagg)] = pd.DataFrame(entries, columns=['model_name', 'fold_idx', 'accuracy'])
```
